# Multilayer Perceptron (MLP) - Ejemplos b√°sicos de uso del algoritmo

Funciones RELU y tanh

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

line = np.linspace(-3, 3, 100)
plt.plot(line, np.tanh(line), label="tanh")
plt.plot(line, np.maximum(line, 0), label="relu")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("relu(x), tanh(x)")
```

# MLP Classifier

Generaci√≥n de datos sint√©ticos:

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)

plt.scatter(X[:, 0], X[:, 1], c=y)
```

Clasificaci√≥n usando MLPClassifier, por defecto se usan 100 nodos:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
random_state=42)
mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X_train, y_train)

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

Experimento con 10 nodos:


```python
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10])
mlp.fit(X_train, y_train)

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

Agregando una segunda capa oculta:

```python
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10, 10])
mlp.fit(X_train, y_train)

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

Por defecto la funci√≥n de activaci√≥n es RELU pero pueden usarse otras, por ejemplo tanh:

```python
mlp = MLPClassifier(solver='lbfgs', activation='tanh', random_state=0, hidden_layer_sizes=[10, 10])
mlp.fit(X_train, y_train)

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)

plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
