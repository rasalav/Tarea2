División del set de train y test:

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(iris, test_size = 0.33)  # in this our main data is split into train and test
# the attribute test_size=0.33 splits the data into 67% and 33% ratio. train=67% and test=33%
print(train.shape)
print(test.shape)

train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]# taking the training data features
train_y = train.Species  # output of our training data
test_X = test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking test data features
test_y = test.Species   #output value of test data
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

Ahora se aplica STD DEV para limpiar el outlier

```python
newiris = irisX[np.abs(irisX.SepalLengthCm-irisX.SepalLengthCm.mean())<=(2*irisX.SepalLengthCm.std())]
```

```python
newiris.head(10)
```

Keep only the ones that are within +3 to -3 standard deviations in the column 'Data'.

```python
newirisfinal = newiris[np.abs(newiris.SepalWidthCm-newiris.SepalWidthCm.mean())<=(2*newiris.SepalWidthCm.std())]
```

```python
newirisfinal.tail(10)
```

```python
newirisfinal.head(10)
```

```python
fig = newirisfinal[newirisfinal.Species=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='orange', label='setosa')
newirisfinal[newirisfinal.Species=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='blue', label='versicolor',ax=fig)
newirisfinal[newirisfinal.Species=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='green', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width X")
fig=plt.gcf()
fig.set_size_inches(8,6)
plt.show()
```
