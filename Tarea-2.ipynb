# **TUTORIAL PANDAS**



Pandas es una Librer√≠a de c√≥digo abierto para el an√°lisis de datos de Python. Es un paquete de datos popular para la Ciencia de los Datos y Machine Learning.

Proporciona herramientas de an√°lisis y manipulaci√≥n de datos de alto rendimiento, utilizando sus potentes estructuras de datos.

El nombre **pandas** proviene de **PAN**el **DA**ta.

Utilizando esta librer√≠a, se puede lograr 5 pasos t√≠picos en el procesamiento y an√°lisis de datos, independientemente del origen de los datos:

1- Cargar.

2- Preparar

3- Manipular

4- Modelar

5- Analizar

La estructura fundamental de pandas es un DATAFRAME. Un DATAFRAME es un aestructura de datos etiquetados bidimensionalmente.

Los pandas DATAFRAME consta de tres componentes principales: los datos, el √≠ndice y las columnas.

Adicionalmente con la estructura PanDas DATAFRAME puedes especificar los nombres de √≠ndice (se refiere a las filas) y columnas (se refiere a columnas).

## Estructuras de los Datos en Pandas ##

a) Arreglos de una dimensi√≥n (**SERIES**)

Tienen un √≠ndice de etiquetas para las columnas.



b) Arreglos de dos dimensiones (**DATAFRAME**)

Tienen un √≠ndice de etiquetas para las filas y un √≠ndice de etiquetas para las columnas.

Estos son los que se emplean para Machine Learning.



c) Arreglos de tres dimensiones (PANELES)

## Caracter√≠sticas Principales de las Librer√≠as de Pandas##

* El objeto DATAFRAME es r√°opido y eficiente con indexaci√≥n predeterminada y personalizada.
* Herramientas para cargar datos en objetos de datos en memoria, desde diferentes formatos de archivos.
* Alineaci√≥n de datos y manejo integrado de datos faltantes.
* Remodelaci√≥n y giro de conjunto de fechas.
* Etiquetado, corte, indexaci√≥n y subconjuntos, de grandes conjuntos de datos.
* Las columnas de una estructura de datos se pueden eliminar o insertar.
* Agruoar por datos para segregaciones o transformaciones.
* Alto rendimiento de fusi√≥n y unio√≥n de datos.
* Funcionalidad de la serie de tiempo.
